# EDA Exploratory Data Analysis

## Gender vs Total Amount

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

## Age Group Distribution

```python
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Total Amount vs Age Group

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group', y= 'Amount', data = sales_age, palette= 'Paired')
```

## Total number of orders from 10 states

```python
state_orders = df.groupby('State')['Orders'].sum().reset_index()
```

```python
top_states = state_orders.sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Orders', data=top_states, palette='viridis')
plt.title('Total Number of Orders from Top 10 States')
plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.tight_layout()
plt.show()
```

##  Marital Status

```python
df
```

```python
marital_status_count = df['Marital_Status'].value_counts().reset_index()
marital_status_count.columns = ['Marital_Status', 'Count']
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Marital_Status', y='Count', data=marital_status_count, palette='magma')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Average Spending per Order by Product Category

```python
product_avg_spent = df.groupby('Product_Category')['Amount'].mean().reset_index()
sns.barplot(x='Product_Category', y='Amount', data=product_avg_spent, palette='Set1')
plt.xticks(rotation=90)
plt.title('Average Spending per Order by Product Category')
plt.show()
```

## Gender Distribution

```python
gender_count = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%', colors=['#ff8569','#57b3ff'], startangle=90)
plt.title('Gender Distribution')
plt.axis('equal') 
plt.show()
```

## State-Wise Orders

```python
state_orders = df.groupby('State')['Orders'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='State', y='Orders', data=state_orders, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Total Orders by State')
plt.show()
```
